- splitting a column using delimiter, then dropping it & renaming new columns

```python
split_col=df['spec_summary'].str.split('|',expand=True)
df=pd.concat([df,split_col],axis=1)
df.head(2)
```

```python
df=df.drop(columns='spec_summary')
```

```python
df=df.rename(columns={0:'Model',1:'Fuel',2:'GearBox',3:'bhp'})
```

```python
df.head(2)
```

- dropped duplicate column

```python
df=df.drop(columns='bhp')
```

- converted values (1 lakh = 1,00,000)

```python
df['total_interest_paid_lakhs']=df['total_interest_paid_lakhs']*100000
df['resale_value_5yrs_lakhs']=df['resale_value_5yrs_lakhs']*100000
df['ownership_cost_5yrs_lakhs']=df['ownership_cost_5yrs_lakhs']*100000
df['net_cost_of_ownership_lakhs']=df['net_cost_of_ownership_lakhs']*100000
df['ex_showroom_price_lakhs']=df['ex_showroom_price_lakhs']*100000
df['onroad_price_lakhs']=df['onroad_price_lakhs']*100000
df['loan_amount_lakhs']=df['loan_amount_lakhs']*100000
```

```python
df.head()
```

- dropped duplicate column

```python
df=df.drop(columns='Model')
df.head()
```

- converted a col to list

```python
df['features_list']=df['features'].str.split(',')
df['features_list']=df['features_list'].apply(lambda x: [i.strip() for i in x] if isinstance(x,list) else x)
```

```python
df=df.drop(columns='features')
df.head(2)
```
