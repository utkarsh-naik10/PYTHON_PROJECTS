```python
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Users\91706\Data Analytics Practice\databases\python\crime data\Crime_Data_from_2020_to_Present.csv')
```

```python
df.isnull().sum() #checking foe null values
```

```python
pd.set_option('display.max_columns',None)
df.head(2)
```

```python
df.shape
```

```python
df.info()
```

```python
df['Vict Sex']=df['Vict Sex'].fillna('Unknown')
df['Vict Descent']=df['Vict Descent'].fillna('Unknown')
df['Weapon Desc']=df['Weapon Desc'].fillna('Unknown')
```

```python
df['Cross Street'].unique()
```

```python
df.drop(columns=['Cross Street'],inplace=True)
```

```python
df.drop(columns=['Crm Cd 3'],inplace=True)
df.drop(columns=['Crm Cd 4'],inplace=True)
```

```python
df.drop(columns='Crm Cd 2',inplace=True)
```

```python
df['Premis Desc']=df['Premis Desc'].fillna('Not Specified')
```

```python
df['Mocodes']=df['Mocodes'].fillna('Unknown')
```

```python
df['Status']=df['Status'].fillna('Unknown')
```

```python
df['DATE OCC']=pd.to_datetime(df['DATE OCC'],format='%m/%d/%Y %I:%M:%S %p')
df['Date Rptd']=pd.to_datetime(df['Date Rptd'],format='%m/%d/%Y %I:%M:%S %p')
```

```python
df['Weapon Used Cd']=df['Weapon Used Cd'].fillna(0)
df['Premis Cd']=df['Premis Cd'].fillna(0)
df['Crm Cd 1']=df['Crm Cd 1'].fillna(0)
```

```python
df.head(2)
```

```python
df['occ_year']=df['DATE OCC'].dt.year
```

```python
df['occ_month']=df['DATE OCC'].dt.strftime('%b')
```

```python
df['occ_date']=df['DATE OCC'].dt.day
```

```python
df['occ_day']=df['DATE OCC'].dt.strftime('%a')
```

```python
df['TIME OCC'] = df['TIME OCC'].astype(str).str.zfill(4)
df['TIME OCC'] = pd.to_datetime(df['TIME OCC'], format='%H%M').dt.time
```

```python
df.head()
```

```python
import os

path = r'C:\Users\91706\Data Analytics Practice\databases\python\crime data'
file = os.path.join(path, 'Crime_Data_from_2020_to_Present.parquet')

os.makedirs(path, exist_ok=True)

df.to_parquet(file, engine='pyarrow', index=False)

print("file saved successfully at:")
print(file)
```
